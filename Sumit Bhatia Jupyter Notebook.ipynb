### Imputing missing values in each column.

```python
# Lets recheck the %age missing values in our data frame now

(application.isnull().mean()*100).sort_values(ascending = False)
```

```python
# Automated approach for imputing the missing values in continuous columns by the median of each column.

for i in application_cont :
    if application[i].isnull().mean() > 0 :
        med = application[i].median()
        application[i] = application[i].fillna(med)
```

```python
# Automated approach for imputing the missing values in categorical columns by the mode of each column.

for i in application_cat :
    if application[i].isnull().mean() > 0 :
        md = application[i].mode()[0]
        application[i] = application[i].fillna(md)
```

```python
# # Let us now recheck the percentage of missing values in each column after imputing.

(application.isnull().mean()*100).sort_values(ascending = False)
```

### Till now our application data set is cleaned off missing values

## Now let us handle outliers in numeric columns

```python
application.head()
```

### Examining column  - AMT_INCOME_TOTAL for outliers

```python
# Automated approach for handling outliers by applying upper bound and lower bound.
for i in application_cont :
    
    Q3 = application[i].quantile(75/100)
    Q1 = application[i].quantile(25/100)
    IQR = Q3-Q1
    UPPERBOUND = Q3+(1.5*IQR)
    LOWERBOUND = Q1-(1.5*IQR)
    application[i] = np.where(application[i]>UPPERBOUND,UPPERBOUND,application[i])
    application[i] = np.where(application[i]<LOWERBOUND,LOWERBOUND,application[i])
```

```python
# Rechecking if our numerical columns are free of outliers.
for i in application_cont :
    sns.boxplot(application[i])
    plt.show()
```

```python
# Binning of AMT_INCOME_TOTAL_BIN into 3 layers as low, medium and high income groups.
application["AMT_INCOME_TOTAL_BIN"] = pd.qcut(application["AMT_INCOME_TOTAL"] , 3 , labels = ["Low" , "Medium" , "High"])
```

```python
application["AMT_INCOME_TOTAL_BIN"].value_counts()
```

```python
# Automated approach for binning in other columns.
l = ['AMT_CREDIT','AMT_ANNUITY','AMT_GOODS_PRICE']
for i in l :
    application[i+"_BIN"] = pd.qcut(application[i] , 3 , labels = ["Low" , "Medium" , "High"])
```

```python
# Creating a column named 'AGE' with age of clients in years.
application["AGE"] = application["DAYS_BIRTH"].apply(lambda x : x/365)
```

```python
# Function to group age of clients into different groups.
def f(x) :
    if 0<x<25 :
        return "Below 25"
    elif 25<x<50 :
        return "25-50"
    elif 50<x<75 :
        return "50-75"
    else :
        return "Above 75"
```

```python
# Creating a column 'AGE_GROUP' with different age groups of clients.
application["AGE_GROUP"] = application["AGE"].apply(f)
```
